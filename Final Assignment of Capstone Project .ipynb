## 1. Introduction

### 1.1 Background

New York City (NYC) is the most populous city in the United States. NYC has also been regarded as the cultural, financial, and media capital of the world, influencing almost every aspect
of business in the world. As one of the world's most tourist visited countries and one of the world's major entertainment centers, most importantly, the diversity and multicuture of NYC provide a lot of
opportunities for business. However, NYC is also one of the most competitive markets in the world. It could be highly risky for opening a new business there, since the cost of maintaining the business is also very high. Therefore, business analysis should always be recommended before any investment. For example, choosing a good location to start the  business could be a critical factor for success. A good understanding of the business 
environment will largely reduce the risk of a falling business. 
The aim of this study is to provide a good location for anyone who wants to open a new Chinese restaurant in New York City through neighbourhood segemention and clustering of NYC area.

### 1.2 Business problem

New York City has attracted a lot of tourists and immigrants every year because of its diversity and activity in business and culture. The international environment has also simultaneously influenced its food culture and cuisines. Chinese cuisine is becoming more and more popular in the United States that Chinese restaurants can be found almost everywhere, especially for the large cities like NYC. Opening a Chinese restaurant could bring a good business opportunity for success, however, cautions are needed to be taken when choosing the restaurant location in order to sustain a long-term success from the high competition.

The analysis in this research will focus on finding a better location to open a Chinese restaurant in Manhattan area where is a borough of New York City. To ensure the new Chinese restaurant attracts enough customers and avoids the high competition from other business at the beginning, the potential neighborhoods based on the number of Chinese restaurants have been explored around the Manhattan area. 

## 2. Data

### 2.1 New York City Neighborhood

NYC has a total of 5 boroughs and 306 neighborhoods. In order to explore, analyze and segment neighbourhoods, a dataset that includes all the 5 boroughs and neighborhoods as well as geographical location of each neighbourhood is needed to provide. 

The dataset can be downloaded directly from the IBM server using wget command or from the official website https://geo.nyu.edu/catalog/nyu_2451_34572

### 2.2 Foursquare API Data

Foursquare API has been used as the data source to provide the venues information for each neighborhood in New York City. 
Foursquare is a location data provider with information about all manner of venues and events within an area of interest.
After obtaining the list of neighborhoods, we then connect to the Foursquare to collect information about all restaurants around every enighborhood within the radius of 1000 meters. 
The data retrieved from Foursquare including the specified venue category, the longitude and latitude of the venue.

The purpose of this analysis was to explore the Manhattan borough in New York City and find a better place to open a Chinese restaurant.
